<b>SIMPLE MANUAL MODEL</b>
<p>here we are going to manually differentiate species based on one parameter that is petal length

````python
def singleFeaturePrediction(petal_length):
  if(petal_length<2.5):
    return 0
  elif(petal_length<4.8):
    return 1
  else:
    return 2
````

````python
df_train.columns
````

````python
x_train[:,2] #way of representing  specific values of a columns in numpy
````

````python
manual = [singleFeaturePrediction(val) for val in x_train[:,2]]
````

````python
manual == y_train
````

````python
manual_modelling = np.mean(manual==y_train)*100
````

````python
print(f"Manual Model Accuracy = {manual_modelling*100:.2f}")
````

Such high accuracy was observed only because the dataset we are using is way to small and can be easily be distinguished .

<h1><b>LOGISTIC REGRESSION</b></h1>
<p>Logistic regression is a statistical method used to analyze and model the relationship between a dependent binary variable (usually denoted by "y" that takes two possible outcomes such as 0 or 1) and one or more independent variables (usually denoted by "x")</p>
<p>
The logistic regression model uses a logistic function (also known as the sigmoid function) to map the input variables to the output probability. The output of the logistic regression model is a number between 0 and 1, which represents the predicted probability of the binary outcome.</P>
